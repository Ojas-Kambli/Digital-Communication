```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
f=1
data=np.array([1,0,1,1,0,1])
```

```python
t=np.arange(0,1,1/500)
x=np.cos(2*pi*f*t)
c1=np.cos((2*pi*f*t)+pi/4)
c2=np.cos((2*pi*f*t)+3*pi/4)
c3=np.cos((2*pi*f*t)+5*pi/4)
c4=np.cos((2*pi*f*t)+7*pi/4)
```

```python
data_odd=[data[i] for i in range(len(data)) if (i%2==0)]
data_even=[data[i] for i in range(len(data)) if (i%2!=0)]
print("The odd bits are : ",data_odd)
print("The even bits are : ",data_even)
```

```python
data_odd=np.repeat(data_odd,2)
data_even=np.repeat(data_even,2)
print("The odd bits are : ",data_odd)
print("The even bits are : ",data_even)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10))
z = 0
lw = 3 # linewidth
fs = 18 # fontsize
for i in range(len(data)-1):
  # Carrier signal
  ax1.plot(t+z, x, linewidth = lw)
  ax1.set_ylabel('Carrier signal', fontsize = fs)
  ax1.grid(which='major', color='k', linestyle='--')
  # QPSK signal
  if data_odd[i+1] == 1:
    if data_even[i] == 1:
      ax2.plot(t+z,c4 , linewidth = lw)
    else:
      ax2.plot(t+z, c3 , linewidth = lw)
  else:
    if data_even[i] == 1:
      ax2.plot(t+z,  c1 , linewidth = lw)
    else:
      ax2.plot(t+z,  c2 , linewidth = lw)
  ax2.set_ylabel('Offset QPSK signal', fontsize = fs)
  ax2.grid(which='major', color='k', linestyle='--')
  z = z + 1
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
  #
```

```python
Ps = 1
Tb = 1
f = np.linspace(-8, 8, 500)
f0 = 2 # carrier frequency

PSD =  Ps * Tb * (np.sinc( (f-f0) * 2 * Tb ))**2
PSD += Ps * Tb * (np.sinc( (f+f0) * 2 * Tb ))**2
plt.plot(f, PSD, 'r', linewidth = 3)
plt.xlabel('Frequency', fontsize = 18)
plt.ylabel('QPSK PSD', fontsize = 18);
```
